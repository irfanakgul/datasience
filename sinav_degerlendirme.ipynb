```python
import numpy as np
import pandas as pd
```

```python
py_mind=r'C:\Users\Toshiba\Desktop\datascience\Pandas\ödev\py_mind.xlsx'
py_opinion=r'C:\Users\Toshiba\Desktop\datascience\Pandas\ödev\py_opinion.xlsx'
py_science=r'C:\Users\Toshiba\Desktop\datascience\Pandas\ödev\py_science.xlsx'
py_sense=r'C:\Users\Toshiba\Desktop\datascience\Pandas\ödev\py_sense.xlsx'
# SINIFLAR AİT DOSYALARA ULAŞMAK İŞİN ADRES DEĞİŞKENLERİ

xlsx = pd.ExcelFile(py_science)
df1=xlsx.parse(4)
# # py_science SINIFINDAN 4.İNDİSLİ SAYFANIN VERİLERİ DataFrame OLARAK df1'E ATIYOR
df1.columns=['a','cevaplar']
df1.drop('a', axis = 1, inplace = True)
# a SUTÜNÜ SİLİNİYOR(SADECE CEVAPLARDAN OLUŞAN BİR SUTÜN ELDE ETMEK İÇİN)

groups=['-']
# HER ÖĞRENCİNİN SINIFININ EKLENECEĞİ 'groups' LİSTESİ
link=[py_sense,py_science,py_opinion,py_mind]
# ADRESLERİN DEĞİŞKENLERİNDEN OLUŞAN 'link'LİSTESİ
group_name=['py_sense','py_science','py_opinion','py_mind']
# 'groups' LİSTESİNE EKLENECEK SINIF İSİMLERİNİN LİSTESİ

for n in range(4) :
    xlsx = pd.ExcelFile(link[n])    
    for i in range(len(xlsx.sheet_names)): 
        #'sheet names' xlsx DOSYASINDAKİ TÜM SAYFALARIN İSİMLERİNİ VERİR
        groups.append(group_name[n])        
        df2=xlsx.parse(i)        
        # 'xlsx.parse(i)' xlsx DOSYASINDAKİ i.İNDİSLİ SAYFAYI DataFrame OLARAK df2'YE ATIYOR
        df1[xlsx.sheet_names[i]]=df2.iloc[:,-1]
        #df1'E KOLON İSMİ 'xlsx.sheet_names[i]' OLARAK df2'NİN SON KOLONU EKLENİYOR  
df1=df1.T
# df1'İN TRANSPOZESİ ALINIYOR
df1['Grup']=groups
# df1'E groups LİSTESİ 'Group' İSİMLİ KOLON OLARAK EKLENİYOR
```

## 1.Tablo

```python
df3=df1.T
# df1'İN TRANSPOZESİ df3'E ATANIYOR
dogru_say=[]
bos=[]
isim=list(df3.columns)
# df3'ÜN KOLON İSİMLERİ(öğrenci ismleri) İSİMLER DEĞİŞKENİ OLARAK ATANIYOR

for i in df3.columns:
    dogru=df3[i].iloc[:20]==df3['cevaplar'].iloc[:20]
    # dF3'ÜN i.COLUMNUNUN 20.SATIRA KADAR VERİLERİ İLE 'cevaplar' COLUMNUNUN 20.SATIRINA KARŞILAŞTIRILIYOR
    # VE dogru İSMLİ PandasSeriLER OLUŞTURULUYOR
    dogru_say.append(dogru[dogru==True].count())
    # dogru SERİSİNDEKİ DOGRU SAYILARI dogru_say İSİMLİ LİSTEYE EKLENİYOR
    bos.append(df3[i].iloc[:20].isna().sum())
    # df3'ÜN i.COLUMNUNUN 20.SATIRA KADAR none VERİLERİ TOPLAMI bos İSİMLİ LİSTEYE EKELNİYOR

df = pd.DataFrame([isim,groups,dogru_say,bos])
# isim,groups,dogru_say,bos LİSTELİERİNDEN df İSİMLİ DataFrame OLUŞTURULUYOR
df=df.T
df['Yanlış']=20-df[2]-df[3]
# df'YE, 20 DEN df'NİN 2 VE 3 İNDİSLİ COLUMNLARI ÇIKARTILARAK OLUŞTURULAN YANLIŞ SAYILARINI 
# VEREN yanlıs İSİMLİ COLUMN EKLENİYOR  
df.columns=['İSİM','SINIF','DOĞRU','BOŞ','YANLIŞ']
# COLUMN İSİMLERİ DEĞİŞTİRİLİYOR
df=df[1:]
# İLK SATIR ATILMIŞ OLDU(drop ile de yapılabilirdi)
df
```

## 2.Sınava giren öğrenci sayısı

```python
print('SINAVA GİREN TOPLAM ÖĞRENCİ SAYISI =',df['İSİM'].count())
```

## 3.Sınıfların doğru-yanlış-boş ortalamaları

```python
dogru_mean=[]
yanlis_mean=[]
bos_mean=[]
for i in group_name:
    # i (i in ['py_sense','py_science','py_opinion','py_mind'] SINIFININ
    dogru_mean.append(df[df['SINIF']==i]['DOĞRU'].mean())
    # DOĞRU ORTALAMASI 
    yanlis_mean.append(df[df['SINIF']==i]['YANLIŞ'].mean())
    # YANLIŞ ORTALAMASI
    bos_mean.append(df[df['SINIF']==i]['BOŞ'].mean())
    # BOS ORTALAMASI
df_mean=pd.DataFrame([group_name,dogru_mean,yanlis_mean,bos_mean],index=['SINIF','DOĞRU ORT.','YANLIŞ ORT.','BOŞ ORT.'])
# OLUŞTURULAN group_name,dogru_mean,yanlis_mean,bos_mean LİSTELERİ İLE DataFrame OLUŞTURDUK
df_mean.T

```

## 4.En basarılı ögrenciler

```python
df['NET']=df['DOĞRU']-df['YANLIŞ']/4
# 'DOĞRU' COLUMNUNDAN 'YANLIŞ' CLOMNUNUN 1/4'Ü ÇIKARILARAK OLUŞTURULAN 'NET' İSİMLİ COLUMN EKLEDİK 
en_iyiler=[]
netler=[]
sinifi=[]
for i in sinif:
    maks=df[df['SINIF']==i]['NET'].max()
    # HERBİR SINIFA AİT EN ÇOK NET 'maks' DEĞİŞKENİNE ATANIR
    a=list(df[(df['SINIF']==i)&(df['NET']==maks)]['İSİM'])
    en_iyiler.extend(a)
    # HERBİR SINIFTA NETİ 'maks'A EŞİT OLAN ÖĞRENCİYİ en_iyiler LİSTESİNE EKLER
    sinifi.extend([i]*len(a))  
    #[i]*len(a) KOMUTU sinif LİSTESİNE NETİ 'maks' OLAN ÖĞRENCİ SAYISI KADAR i SINIFINI 
    netler.extend([maks]*len(a))
    # VE maks NETİ EKLER
df_best=pd.DataFrame([sinifi,en_iyiler,netler],index=['SINIF','İSİM','NET'])
# sinifi,en_iyiler,netler LİSTELERİ İLE DataFrame OLUŞTURDUK
df_best.T
   
    
```
